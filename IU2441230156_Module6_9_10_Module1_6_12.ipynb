#10. Print total number of Even numbers and Total number of Odd Numbers.
Example:
1) function(21), function(22),function(35),function(36),function(40).
2) 21 is odd, 22 is even, 35 is odd,36 is even, 40 is even.
3) Even={22,36,40} Odd={21,35}
total number of even = 3, total number of odd=2


```python
def count_even_odd(numbers):
    evens = [num for num in numbers if num % 2 == 0]
    odds = [num for num in numbers if num % 2 != 0]

    # Print individual classifications
    for num in numbers:
        status = "even" if num % 2 == 0 else "odd"
        print(f"{num} is {status}", end=", " if num != numbers[-1] else ".\n")

    # Print the grouped sets and totals
    print(f"Even={{{','.join(map(str, evens))}}} Odd={{{','.join(map(str, odds))}}}")
    print(f"total number of even = {len(evens)}, total number of odd = {len(odds)}")

# Example usage:
count_even_odd([21, 22, 35, 36, 40])
```

#Do following:
#1) Define and Call user defined function for n numbers one by one.
#2) Check if each element is Even or Odd and Print it. (Use Class Variable)
#3) Print List of even numbers and odd numbers.

```python
class NumberProcessor:
    even_numbers = []
    odd_numbers = []

    def check_and_store(self, number):
        if number % 2 == 0:
            print(f"Element {number} is Even")

            NumberProcessor.even_numbers.append(number)
        else:
            print(f"Element {number} is Odd")
            NumberProcessor.odd_numbers.append(number)
numbers_list = [12, 7, 23, 44, 90, 15]
processor = NumberProcessor()
print("--- Individual Element Check ---")
for num in numbers_list:
    processor.check_and_store(num)
print("\n--- Final Results Summary ---")
print("List of even numbers:", NumberProcessor.even_numbers)
print("List of odd numbers :", NumberProcessor.odd_numbers)
```
