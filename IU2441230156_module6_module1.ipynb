7. Write a menu driven program to implement the following methods on set. (1) Create (2) Update particular element of set (3) Append to          the set (4) Delete whole set (5) Delete particular element (6) Sort the set (7) Find length

```python
s = set()
print("\n----- MENU -----")
print("1. Create Set")
print("2. Update Particular Element")
print("3. Append to Set")
print("4. Delete Whole Set")
print("5. Delete Particular Element")
print("6. Sort the Set")
print("7. Find Length")
print("8. Exit")
while True:

    choice = int(input("Enter your choice: "))
    if choice == 1:
        n = int(input("Enter number of elements: "))
        s = set()
        for i in range(n):
            x = input("Enter element: ")
            s.add(x)
        print("Set is:", s)

    elif choice == 2:
        old = input("Enter element to update: ")
        if old in s:
            new = input("Enter new element: ")
            s.remove(old)
            s.add(new)
            print("Updated Set:", s)
        else:
            print("Element not found.")

    elif choice == 3:
        x = input("Enter element to append: ")
        s.add(x)
        print("Set after appending:", s)

    elif choice == 4:
        s.clear()
        print("Whole set deleted.")

    elif choice == 5:
        x = input("Enter element to delete: ")
        if x in s:
            s.remove(x)
            print("Set after deletion:", s)
        else:
            print("Element not found.")

    elif choice == 6:
        temp = list(s)
        temp.sort()
        print("Sorted Set:", temp)

    elif choice == 7:
        print("Length of set is:", len(s))

    elif choice == 8:
        print("Program Ended.")
        break

    else:
        print("Invalid Choice.")
```

#3 Matrix multiplication and vector matrix

```python
import numpy as np
# -------- Vector Operations --------
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
print("Vector 1:", v1)
print("Vector 2:", v2)
# Vector Addition
print("Vector Addition:", v1 + v2)
# Vector Multiplication (Element-wise)
print("Vector Multiplication:", v1 * v2)
# -------- Matrix Operations --------
m1 = np.array([[1,2],[3,4]])
m2 = np.array([[5,6],[7,8]])
print("\nMatrix 1:")
print(m1)
print("\nMatrix 2:")
print(m2)
# Matrix Addition
print("\nMatrix Addition:")
print(m1 + m2)
# Matrix Multiplication
print("\nMatrix Multiplication:")
print(np.dot(m1, m2))
```

# Practical based on Strings (Length finding, change specific character,
# palindrome,concatenation)
# String Operations

```python

# Input Strings
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")
# 1. Find Length
print("Length of first string:" ,len(str1))
# 2. Change Specific Character
index = int(input("Enter index to change: "))
new_char = input("Enter new character: ")
if 0 <= index < len(str1):
  str1 = str1[:index] + new_char + str1[index + 1:]
  print("Updated String:", str1)
else:
  print("Invalid index!")
# 3. Palindrome Check
if str1 == str1[::-1]:
  print("The string is a Palindrome.")
else:
  print("The string is Not a Palindrome.")
# 4. Concatenation
result = str1 + str2
print("Concatenated String:", result)
```

```python
# Menu Driven Program for Python Operators

print("\n----- MENU -----")
print("1. Arithmetic Operators")
print("2. Comparison Operators")
print("3. Logical Operators")
print("4. Assignment Operators")
print("5. Exit")

while True:
    choice = int(input("Enter your choice: "))

    if choice == 1:
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))

        print("Addition:", a + b)
        print("Subtraction:", a - b)
        print("Multiplication:", a * b)
        print("Division:", a / b)
        print("Modulus:", a % b)
        print("Exponent:", a ** b)
        print("Floor Division:", a // b)

    elif choice == 2:
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))

        print("a == b :", a == b)
        print("a != b :", a != b)
        print("a > b :", a > b)
        print("a < b :", a < b)
        print("a >= b :", a >= b)
        print("a <= b :", a <= b)

    elif choice == 3:
        x = bool(int(input("Enter first value (1=True, 0=False): ")))
        y = bool(int(input("Enter second value (1=True, 0=False): ")))

        print("AND:", x and y)
        print("OR :", x or y)
        print("NOT x:", not x)

    elif choice == 4:
        a = int(input("Enter a number: "))

        print("Original Value:", a)
        a += 5
        print("After += 5:", a)
        a -= 2
        print("After -= 2:", a)
        a *= 3
        print("After *= 3:", a)
        a /= 2
        print("After /= 2:", a)

    elif choice == 5:
        print("Program Ended.")
        break

    else:
        print("Invalid Choice")
```
