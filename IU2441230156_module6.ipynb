6. Write a menu driven program to implement the following methods on List. (1) Create (2) Update particular element of list (3) Append to        the list (4) Delete whole list (5) Delete particular element (6) Sort the list (7) Find length


```python
my_list = []
print("=====Menu=====")
print("1 Create")
print("2. Update an element")
print("3. Append to the list")
print("4. Delete whole list")
print("5. Delete an particular element")
print("6. Sort the list")
print("7. Find the length")
print("8. exit")
while(True):
  choice = int(input("Enter a number(1-8):"))
  if choice == 1:
    my_list = input("Enter elements seperated by spaces").split()
    print(my_list)
  elif choice == 2:
    n = int(input("Enter the index of number to be updated"))
    if n<=8:
      my_list[n] = input("Enter the new element")
      print(my_list)
    else:
      print("Invalid input")
      continue
  elif choice == 3:
    my_list.append(input("Enter the number :"))
  elif choice == 4:
    my_list.clear()
    print(my_list)
  elif choice == 5:
    val = input("Value to remove: ")
    if val in my_list: my_list.remove(val)
    print(my_list)
  elif choice == 6:
    my_list.sort()
    print(my_list)
  elif choice == 7:
    print("length: " ,len(my_list))
  elif choice == 8:
    break

```

```python
dictionary = {}
print("\n(1) Create")
print("(2) Update particular element of dictionary")
print("(3) Append to the dictionary")
print("(4) Delete whole dictionary")
print("(5) Delete particular element")
print("(6) Sort the dictionary")
print("(7) Find length")
print("(8) Exit")
while True:
    choice = int(input("\nEnter your choice: "))

    if choice == 1:
        n = int(input("Enter number of elements to add: "))
        dictionary = {}
        for i in range(n):
            k = input("Enter key: ")
            v = input("Enter value: ")
            dictionary[k] = v
        print("Current Dictionary:", dictionary)

    elif choice == 2:
        k = input("Enter key to update: ")
        if k in dictionary:
            v = input("Enter new value: ")
            dictionary[k] = v
            print("Current Dictionary:", dictionary)
        else:
            print("Key not found.")

    elif choice == 3:
        k = input("Enter key to append: ")
        v = input("Enter value: ")
        dictionary[k] = v
        print("Current Dictionary:", dictionary)

    elif choice == 4:
        dictionary.clear()
        print("Current Dictionary:", dictionary)

    elif choice == 5:
        k = input("Enter key to delete: ")
        if k in dictionary:
            del dictionary[k]
            print("Current Dictionary:", dictionary)
        else:
            print("Key not found.")

    elif choice == 6:
        dictionary = dict(sorted(dictionary.items()))
        print("Sorted Dictionary:", dictionary)

    elif choice == 7:
        print("Length of Dictionary:", len(dictionary))

    elif choice == 8:
        break

    else:
        print("Invalid choice")
```
